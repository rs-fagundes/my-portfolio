```python
import pandas as pd
import matplotlib.pyplot as plt
import wbdata
```

```python
# Define the query parameters
indicator = {'SL.EMP.TOTL.SP.ZS': 'Employment Rate (% of Total Population)'}
country = 'PT'

# Fetch data from the World Bank API
data = wbdata.get_dataframe(indicator, country=country)
```

```python
# Sort data by year (index) in ascending order
data_sorted = data.sort_index()

# Plotting the employment trend with years in ascending order
plt.figure(figsize=(12, 6))
data_sorted.plot(legend=None)
plt.title('Employment Trend in Portugal')
plt.xlabel('Year')
plt.ylabel('Employment Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
