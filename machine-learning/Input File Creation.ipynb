```python
import warnings
import pandas as pd


pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
warnings.filterwarnings('ignore')
```

```python
df_calendar = pd.read_csv('csv/Calendar.csv', parse_dates=['Date'])

df_orders = pd.read_csv('csv/Orders.csv',parse_dates=['Day'])

df_orders.rename(columns={'Day':'Date'},inplace=True)

df_orders_calendar = pd.merge(df_orders,df_calendar, on='Date', how='left')

df_orders_calendar_grouped = df_orders_calendar.groupby(['Fiscal Week Year'])['Total Units'].sum().reset_index()
```

```python
df_orders_calendar_grouped.head()
```

```python
folder_path = r'C:\Users\Rafael_Fagundes\Downloads\compiled_data_v2.csv'

df_delve = pd.read_csv(folder_path)

df_delve_grouped = df_delve[df_delve['Country'] == 'United States'].groupby(['Fiscal Week Year','Display Dell Vehicle Mapped'])['Spend'].sum().reset_index()

df_delve_grouped_pivoted = df_delve_grouped.pivot(index='Fiscal Week Year', columns='Display Dell Vehicle Mapped')['Spend'].reset_index()

df_final = pd.merge(df_delve_grouped_pivoted, df_orders_calendar_grouped, on='Fiscal Week Year', how='left')

df_final.fillna(0,inplace=True)
```
